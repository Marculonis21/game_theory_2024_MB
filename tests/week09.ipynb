```python
import matplotlib.pyplot as plt
import numpy as np

import efg
import week6
import week8
import week9
```

```python
np.set_printoptions(precision=8)
```

```python
# In all examples, all CFR variants are initialized using uniform strategies
# You can compare the final strategies with optimal strategies listed on Wikipedia
# https://en.wikipedia.org/wiki/Kuhn_poker
kuhn_poker = efg.KuhnPoker()

strategies = week8.counterfactual_regret_minimization(kuhn_poker, epochs=1000)
exploitability_values = week6.plot_exploitability(kuhn_poker, strategies, 'CFR')
print(f'Final exploitability value of CFR: {exploitability_values[-1]:.8f}', end='\n\n')

strategies = week9.cfr_br(kuhn_poker, epochs=1000)
exploitability_values = week6.plot_exploitability(kuhn_poker, strategies, 'CFR-BR')
print(f'Final exploitability value of CFR-BR: {exploitability_values[-1]:.8f}', end='\n\n')

strategies = week9.cfr_plus(kuhn_poker, epochs=1000)
exploitability_values = week6.plot_exploitability(kuhn_poker, strategies, 'CFR+')
print(f'Final exploitability value of CFR+: {exploitability_values[-1]:.8f}', end='\n\n')

plt.legend()
plt.show()
```

```python
# This implements the Poker-like game tree presented in the lecture
lecture_poker = efg.LecturePoker()

strategies = week8.counterfactual_regret_minimization(lecture_poker, epochs=500)
exploitability_values = week6.plot_exploitability(lecture_poker, strategies, 'CFR')
print(f'Final exploitability value of CFR: {exploitability_values[-1]:.8f}', end='\n\n')

strategies = week9.cfr_br(lecture_poker, epochs=500)
exploitability_values = week6.plot_exploitability(lecture_poker, strategies, 'CFR-BR')
print(f'Final exploitability value of CFR-BR: {exploitability_values[-1]:.8f}', end='\n\n')

strategies = week9.cfr_plus(lecture_poker, epochs=500)
exploitability_values = week6.plot_exploitability(lecture_poker, strategies, 'CFR+')
print(f'Final exploitability value of CFR+: {exploitability_values[-1]:.8f}', end='\n\n')

plt.legend()
plt.show()
```

```python
# This is a sequential version of Rock Paper Scissors where
# the second player *can* observe actions of the first player
rock_paper_scissors = efg.RockPaperScissors()

strategies = week8.counterfactual_regret_minimization(rock_paper_scissors, epochs=250)
exploitability_values = week6.plot_exploitability(rock_paper_scissors, strategies, 'CFR')
print(f'Final exploitability value of CFR: {exploitability_values[-1]:.8f}', end='\n\n')

strategies = week9.cfr_br(rock_paper_scissors, epochs=250)
exploitability_values = week6.plot_exploitability(rock_paper_scissors, strategies, 'CFR-BR')
print(f'Final exploitability value of CFR-BR: {exploitability_values[-1]:.8f}', end='\n\n')

strategies = week9.cfr_plus(rock_paper_scissors, epochs=250)
exploitability_values = week6.plot_exploitability(rock_paper_scissors, strategies, 'CFR+')
print(f'Final exploitability value of CFR+: {exploitability_values[-1]:.8f}', end='\n\n')

plt.legend()
plt.show()
```
